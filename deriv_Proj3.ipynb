```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cube_deriv(a,b,c,d):
    if (a or b or c or d) < -10:
        return "Please enter values more than -11"
    elif (a or b or c or d) > 10:
        return "Please enter values less than 11"
    else:
        a_deriv = a * 3
        b_deriv = b * 2
        c_deriv = c * 1
        d_deriv = 0
        x = np.linspace(-10,10,1000)
        funct = a*(x**3) + b*(x**2) + c*(x**1) + d
        deriv_funct = a_deriv*(x**2) + b_deriv*(x**1) + c_deriv + d_deriv
        plt.title("Plot of f(x) and f'(x)")
        plt.plot(x,funct,label = "f(x)")
        plt.plot(x,deriv_funct,label = "f'(x)")
        plt.legend()
        plt.show()
```

```python
cube_deriv(3,-2,0,5)
```

```python
cube_deriv(-7,3,-9,10)
```

```python
cube_deriv(-20, -20, 20, 20)
```
