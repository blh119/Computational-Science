###Subway Rides###

This program was created to see if it would be worth it for a MTA patron to either pay for each one way ride or buy the weekly pass.

The weekly pass is 33 dollars and each ride is 2.75. This mean that a weekly pass is worth 12 rides. This mean that if you are going to get 12 or more passes you should buy the pass. Any less than that should buy the rides.

The program only accepts the number of rides to be between 1 and 30 and messages for people who do know meet the number of rides to buy a weekly pass and for those who do. 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ride_cost = 2.75
week_pass = 33.0
```

```python
def min_sub_cost(n):
    rides = ride_cost * n
    if n < 1 or n > 30:
        print("Number of rides can only be between 1 and 30...")
    elif rides < week_pass:
        print("No need to buy the pass...Buy the rides")
    else:
        print("Buy the pass...You get " + str(n) + " rides and more!")
    
```

```python
min_sub_cost(2)
```

```python
min_sub_cost(11)
```

```python
min_sub_cost(12)
```

```python
min_sub_cost(-1)
```

```python
min_sub_cost(31)
```

```python
number_of_rides = np.arange(1,31)
number_of_rides
```

```python
prices_of_rides = number_of_rides * 2.75
prices_of_rides
```

```python
plt.scatter(prices_of_rides, number_of_rides, label = "Ride Price Over Time")
plt.axhline(y=12, color='g', linestyle='--')
plt.ylabel("Number of Rides, n")
plt.xlabel("Price of Rides, $")
plt.title("Price of Rides vs Number of Rides")
```
