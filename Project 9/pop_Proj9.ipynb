In this problem we were given the birth and death rates of female sheep in 1965. We are tasked with finding population at 1975. 

We did this my making a list that stores the number of sheep in each age group. In a for loop for each year in the simulation, we took the value at the end of the list and then used those values to apply the birth and death rates for the new year. 

After the formula is applied for the ten years we stored the values of the last year in a numpy array and summed the last values to get the total population in 1975. 

Able to give a graphical representation for the years that we are tracking the population.

We see that from ages 0 - 6 the female sheep population is growing at almost a exponential rate, especially in the later years. We see that the population from 6 - 12 is a little more choatic. The population 10 - 12 remains pretty constant, While there seems to be a rapid increase in the population ages 6 - 10 

```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
def sheep(n):
    A = [120] #Age 0 - 1
    B = [156] #Age 1 - 2
    C = [175] #Age 2 - 3
    D = [201] #Age 3 - 4
    E = [186] #Age 4 - 5
    F = [153] #Age 5 - 6
    G = [134] #Age 6 - 7
    H = [126] #Age 7 - 8
    I = [98]  #Age 8 - 9
    J = [76]  #Age 9 - 10
    K = [52]  #Age 10 - 11
    L = [22]  #Age 11 - 12
    for i in range(n):
        #Old Equations
        A_old = A[-1] #Age 0 - 1
        B_old = B[-1] #Age 1 - 2
        C_old = C[-1] #Age 2 - 3
        D_old = D[-1] #Age 3 - 4
        E_old = E[-1] #Age 4 - 5
        F_old = F[-1] #Age 5 - 6
        G_old = G[-1] #Age 6 - 7
        H_old = H[-1] #Age 7 - 8
        I_old = I[-1] #Age 8 - 9
        J_old = J[-1] #Age 9 - 10
        K_old = K[-1] #Age 10 - 11
        L_old = L[-1] #Age 11 - 12
        #New Equations
        A_new = B_old*(.045) + C_old*(.391) + D_old*(.472) + E_old*(.484) + F_old*(.546) + G_old*(.543) + H_old*(.502) + I_old*(.468) + J_old*(.459) + K_old*(.433) + L_old*(.421)
        B_new = A_old*(.845)
        C_new = B_old*(.975)
        D_new = C_old*(.965)
        E_new = D_old*(.950)
        F_new = E_old*(.926)
        G_new = F_old*(.895)
        H_new = G_old*(.850)
        I_new = H_old*(.786)
        J_new = I_old*(.691)
        K_new = J_old*(.561)
        L_new = K_old*(.370)
        #Add to list
        A.append(A_new)
        B.append(B_new)
        C.append(C_new)
        D.append(D_new)
        E.append(E_new)
        F.append(F_new)
        G.append(G_new)
        H.append(H_new)
        I.append(I_new)
        J.append(J_new)
        K.append(K_new)
        L.append(L_new)
    return A, B, C, D, E, F, G, H, I, J, K, L   
```

```python
a, b, c, d, e, f, g, h, i, j, k, l = sheep(10)
```

```python
pop = np.array([a[-1], b[-1], c[-1], d[-1], e[-1], f[-1], g[-1], h[-1], i[-1], j[-1], k[-1], l[-1]])
```

```python
sum(pop)
```
